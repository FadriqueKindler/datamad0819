```python
df_sharks_renamecols = df_sharks_clean.rename(columns={'Case Number':'Casenumber',
                            'Case Number.1':'Casenumber1',
                           "Case Number.2" : "Casenumber2"})
```

```python
list_casenumber = list(df_sharks_renamecols.Casenumber.values)

list_casenumber1 = list(df_sharks_renamecols.Casenumber1.values)

list_casenumber2 = list(df_sharks_renamecols.Casenumber2.values)
```

```python
collections.Counter(list_casenumber) == collections.Counter(list_casenumber1) == collections.Counter(list_casenumber2)
```

```python
def differences(x, y):
    if len(x) != len(y):
        raise ValueError("Lists of different length.")
    return sum(i != j for i, j in zip(x, y))

print(str(differences(list_casenumber ,list_casenumber1 )) + " differences between Case Number and Case Number.1")
print(str(differences(list_casenumber ,list_casenumber2 )) + " differences between Case Number and Case Number.2")
print(str(differences(list_casenumber2 ,list_casenumber1 )) + " differences between Case Number.2 and Case Number.1")
```
