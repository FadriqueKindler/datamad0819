```python
import requests
import pandas as pd
import os
import json
from bs4 import BeautifulSoup
```

```python
# IMPORT THE DATASET NBA Players 

df_players = pd.read_csv("../your-code/DATA/nba-players-stats/players.csv")
df_players.head()
```

```python
# DROP UNWANTED COLUMNS

df_players.drop(columns=["Unnamed: 0","birth_city","birth_state","born"],inplace=True)

# RENAME COLUMNS 

df_players = df_players.rename(columns={"height": "Height","weight": "Weight","collage": "College"})


df_players.head()
```

```python
# IMPORT THE DATASET NBA Player Data 

df_player_data = pd.read_csv("../your-code/DATA/nba-players-stats/player_data.csv")
df_player_data.head()
```

```python
# DROP UNWANTED COLUMNS

df_player_data.drop(columns=["height","weight"],inplace=True)

# RENAME COLUMNS TO MATCH FOR MERGING

df_player_data = df_player_data.rename(columns={"name": "Player","year_start": "Rookie","year_end": "Retires","birth_date": "Birth Date","position": "Position", "college":"College"})


df_player_data.head()
```

```python
# IMPORT THE DATASET NBA Season Stats 

df_season_stats = pd.read_csv("../your-code/DATA/nba-players-stats/Seasons_Stats.csv")
df_season_stats.head()
```

```python
# CHECK NUMBER OF MISSING VALUES IN DATAFRAMES

total_values = len(df_season_stats)
print(total_values)

df_season_stats.isnull().sum(axis = 0)
```

```python
# DROP UNWANTED COLUMNS

df_season_stats.drop(columns=["Unnamed: 0","blanl","blank2","Age","Year",],inplace=True)

# DELETE ROWS WHERE PLAYER IS MISSING

df_season_stats = df_season_stats[df_season_stats.Player.notnull()]
df_season_stats.isnull().sum(axis = 0)

# KEEP ONLY ROWS WITH AT LEAST 50 COLUMNS 

df_season_stats.dropna(thresh=50)

# RENAME COLUMNS 

#df_season_stats = dfseason_stats.rename(columns={"year_start": "Rookie","year_end": "Retires","birth_date": "Birth Date","position": "Position"})


df_season_stats.head()
```

```python
# MERGING ALL DATAFRAMES INTO 1

df_players_full = df_players.merge(df_player_data, on="Player")
complete_df = df_players_full.merge(df_season_stats, on="Player")
complete_df.head()
```

```python
#DROPING DUPLICATE ROWS. KEEPING FIRST VALUE AS UNIQUE

complete_df.drop_duplicates(subset ="Player", 
                     keep = "first", inplace = True)
complete_df
```
